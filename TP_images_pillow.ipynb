1)	On souhaite appliquer un filtre rouge.
On parcourt chacun des pixels de l’image (double boucle en faisant varier l’abscisse puis l’ordonnée).

Pour chaque pixel, on récupère son niveau de rouge r, son niveau de vert v et son niveau de bleu b.
Le rouge r est porté à son maximum 255 mais v et b ne sont pas modifiés.


```python
def rouge(img):
    largeur,hauteur=img.size
    for x in range(largeur):
        for y in range(hauteur):
            r,v,b=img.getpixel((x,y))
            img.putpixel((x,y),(255,v,b))
    return img
```

```python
rouge(monalisa)
```
